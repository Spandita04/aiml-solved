```python
#i
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.DESCR)
```

```python
#ii
print(digits.data.shape)
print(digits.target.shape)
```

```python
#iii
print(digits.data[91])
type(digits.data[91])
```

```python
#iv
import numpy as np 
np.reshape(digits.data[91],(8,8))
```

```python
#v
import matplotlib.pyplot as plt
plt.imshow(np.reshape(digits.data[91],(8,8)), cmap='gray')
```

```python
#vi
from sklearn.model_selection import train_test_split
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(digits.data,digits.target,test_size=0.25)
```

```python
#vii
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth=20)
classifier.fit(x_train_1, y_train_1)
y_pred=classifier.predict(x_train_1)
```

```python
#viii
import numpy as np
from sklearn.metrics import r2_score
r2_score(y_train_1,y_pred)
```

```python
#ix
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train_1, y_train_1)
y_pred=classifier.predict(x_train_1)
r2_score(y_train_1,y_pred)
```
