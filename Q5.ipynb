```python
#i
from sklearn.datasets import make_blobs
dataset = make_blobs(n_samples = 400, centers = 4, n_features = 3, cluster_std = 1.5, random_state = 30)
print(dataset)
```

```python
#ii
import matplotlib.pyplot as plt
plt.scatter(dataset[0][:,0],dataset[0][:,1])
```

```python
#iii
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
#iv
y_pred=kmeans.fit_predict(dataset[0])
print(y_pred)
clusters = kmeans.cluster_centers_
print(clusters)
```

```python
#v
plt.scatter(dataset[0][y_pred==0,0], dataset[0][y_pred==0,1], s=80, color='orange')
plt.scatter(dataset[0][y_pred==1,0], dataset[0][y_pred==1,1], s=80, color='cyan')
plt.scatter(dataset[0][y_pred==2,0], dataset[0][y_pred==2,1], s=80, color='yellow')
plt.scatter(dataset[0][y_pred==3,0], dataset[0][y_pred==3,1], s=80, color='pink')
plt.scatter(clusters[0][0], clusters[0][1], marker="+", s=80, color="black")
plt.scatter(clusters[1][0], clusters[1][1], marker="+", s=80, color="black")
plt.scatter(clusters[2][0], clusters[2][1], marker="+", s=80, color="black")
plt.scatter(clusters[3][0], clusters[3][1], marker="+", s=80, color="black")
```
