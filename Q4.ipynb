```python
#i
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
#ii
from sklearn.model_selection import train_test_split
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(digits.data,digits.target,test_size=0.25)
```

```python
#iii
import numpy as np
np.arange(len(x_train_1))
```

```python
#iv
y_train_nonlabel=np.copy(y_train_1)
y_train_nonlabel[np.arange(len(x_train_1))[280:]]=-1
```

```python
#v
from sklearn.semi_supervised import LabelPropagation
train_model=LabelPropagation()
train_model.fit(x_train_1,y_train_1)
```

```python
#vi
from sklearn.metrics import r2_score
y_pred = train_model.predict(x_test_1)
r2_score(y_test_1,y_pred)
```

```python
#vii
train_model2=LabelPropagation(gamma=0.2)
train_model2.fit(x_train_1,y_train_1)
```

```python
#viii
y_pred = train_model2.predict(x_test_1)
r2_score(y_test_1,y_pred)
```
